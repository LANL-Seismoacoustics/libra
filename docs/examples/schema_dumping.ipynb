## Writing to a Database

```python
import os

import sqlalchemy
from sqlalchemy.orm import Session

from libra import Schema
from libra.util import DatabaseSettings

# Go ahead and load in from a pre-defined yaml file
kbcore = Schema('NNSA KB Core').load('data/kbcore.yaml')

database_path = 'data/examples.db'

# Cleanup if needed
if os.path.exists(database_path):
    os.remove(database_path)

engine = sqlalchemy.create_engine(f'sqlite:///{database_path}')
session = Session(engine)

settings = DatabaseSettings(
    engine = engine,
    create_tables = True,
    author = 'NNSA'
)

kbcore.dump(settings)
```
